```python
mean=df['PREVAILING_WAGE'].mean()
```

```python
most_common2 = df['EMPLOYER_NAME'].mode()[0]
```

```python
most_common3 = df['JOB_TITLE'].mode()[0]
```

```python
most_common4 = df['YEAR'].mode()[0]
```

```python
df.fillna({'PREVAILING_WAGE':mean},inplace= True)
```

```python
df.fillna({'EMPLOYER_NAME': most_common2}, inplace=True)
```

```python
df.fillna({'JOB_TITLE': most_common3}, inplace=True)
```

```python
df.fillna({'YEAR': most_common4}, inplace=True)
```

```python
df['CASE_STATUS'] = df['CASE_STATUS'].fillna("WITHDRAWN")
```

```python
df['FULL_TIME_POSITION'] = df['FULL_TIME_POSITION'].fillna("N")
```

# Top 10 Hiring Companies: 

```python
top_10=df[['EMPLOYER_NAME']].value_counts().head(10).to_frame().reset_index()
```

```python
top_10
```

```python
import plotly.graph_objs as go
data=top_10
trace = go.Bar(x=data['EMPLOYER_NAME'],y=data['count'],marker=dict(color='green'))
layout = go.Layout(title='Simple Bar Graph', yaxis=dict(title='Hiring'), xaxis=dict(title='Companies'))
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

# Distribution of Prevailing Wages

```python
distribution=df.groupby('EMPLOYER_NAME')['PREVAILING_WAGE'].sum().sort_values(ascending=False).to_frame().reset_index().head(10)
```

```python
distribution
```

```python
import plotly.express as px
import pandas as pd

dS = distribution

fig = px.scatter(dS, y='PREVAILING_WAGE',x='EMPLOYER_NAME', title='Box Plot')
fig.show()
```

# Average Prevailing Wage by Employer

```python
wages_average = df.groupby('EMPLOYER_NAME')['PREVAILING_WAGE'].mean().sort_values(ascending=False).head(10).to_frame().reset_index().rename(columns={'PREVAILING_WAGE': 'Average_Wage'})
```

```python
wages_average
```

```python
import plotly.express as px

fig = px.bar(
    wages_average,
    x= 'Average_Wage',
    y='EMPLOYER_NAME',
    orientation='h',
    title='Top 10 Companies by Average Prevailing Wage',
    labels={'PREVAILING_WAGE': 'Average Wage', 'EMPLOYER_NAME': 'Company'},
    color='Average_Wage',
    color_continuous_scale='reds'
)

fig.update_layout(yaxis=dict(categoryorder='total ascending'))
fig.show()
```

# Top 10 Companies Offering Highest Salaries:

```python
top_10= df.groupby('EMPLOYER_NAME')['PREVAILING_WAGE'].max().head(10).sort_values(ascending= False).to_frame().reset_index().rename(columns={'PREVAILING_WAGE':'Highest salaries'})
```

```python
top_10
```

```python
fig = px.histogram(
   top_10,
    x= 'Highest salaries',
    y='EMPLOYER_NAME',
    orientation='h',
    title='Top 10 Companies by Highest salaries',
    labels={'PREVAILING_WAGE': 'Highest salaries', 'EMPLOYER_NAME': 'Company'},
    color='Highest salaries',
   
)

fig.update_layout(yaxis=dict(categoryorder='total ascending'))
fig.show()
```

#  Denied Petitions Over the Years:

```python
dp=(df['CASE_STATUS'] == 'DENIED').groupby(df['YEAR']).sum().to_frame().reset_index()
dp.columns=['YEARS','Denied petition']
dp
```
